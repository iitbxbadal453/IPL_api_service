```python
team1 = 'Rajasthan Royals'
team2 = 'Royal Challengers Bangalore'
```

```python
df = matches[((matches['Team1'] == team1) & (matches['Team2'] == team2)) | ((matches['Team1'] == team2) & (matches['Team2'] == team1))]
df.head()
```

```python
matches_won_by_team1 = df['WinningTeam'].value_counts()[team1]
matches_won_by_team2 = df['WinningTeam'].value_counts()[team2]
draws = total - matches_won_by_team1 - matches_won_by_team2
```

```python
def team1vsteam2(team, team2):
    df = matches[((matches['Team1'] == team) & (matches['Team2'] == team2)) | (
                (matches['Team2'] == team) & (matches['Team1'] == team2))].copy()
    mp = df.shape[0]
    won = df[df.WinningTeam == team].shape[0]
    nr = df[df.WinningTeam.isnull()].shape[0]
    loss = mp - won - nr

    return {'matchesplayed': mp,
            'won': won,
            'loss': loss,
            'noResult': nr}
```

```python
team1vsteam2('Royal Challengers Bangalore', 'Chennai Super Kings')
```
